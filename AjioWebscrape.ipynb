```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Initialize WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--disable-notifications")
driver = webdriver.Chrome(options=options)
```

```python
# Open the target webpage
driver.get("https://www.ajio.com/men-jackets-coats/c/830216010")
driver.maximize_window()
```

```python
import pandas as pd

# Initialize a list to store product details
product_list = []

# Function to scrape product details
def scrape_product_details():
    while True:
        try:
            # Wait for products to load
            WebDriverWait(driver, 10).until(
                EC.presence_of_element_located((By.CLASS_NAME, "brand"))
            )

            # Locate all product containers
            products = driver.find_elements(By.XPATH, "//div[contains(@class, 'product')]")

            for product in products:
                try:
                    # Extract details for each product
                    brand = product.find_element(By.CLASS_NAME, "brand").find_element(By.TAG_NAME, "strong").text
                    name = product.find_element(By.CLASS_NAME, "nameCls").text
                    price = product.find_element(By.XPATH, ".//span[contains(@class, 'price')]/strong").text
                    discount = product.find_element(By.CLASS_NAME, "discount").text if product.find_elements(By.CLASS_NAME, "discount") else "No discount"

                    # Print the details
                    print(f"Brand: {brand}, Name: {name}, Price: {price}, Discount: {discount}")

                    # Store the details in the product list
                    product_list.append({
                        "Brand": brand,
                        "Name": name,
                        "Price": price,
                        "Discount": discount
                    })
                except Exception as e:
                    print(f"Error extracting product details: {e}")

            # Try to click on the "Next" button
            try:
                next_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, "//a[@data-testid='pagination-next']"))
                )
                next_button.click()
                time.sleep(2)  # Wait for the next page to load
            except Exception as e:
                print(f"No more pages to load or unable to click 'Next': {e}")
                break  # Exit the loop if no more pages

        except Exception as e:
            print(f"Error while scraping: {e}")
            break  # Exit the loop on major error

# Main function
scrape_product_details()

# Save the product details to a CSV file
df = pd.DataFrame(product_list)
df.to_csv('products.csv', index=False)

print("Product details saved to products.csv")

# Close the driver
driver.quit()
```
